```python
data_hex = "00000000A714418801222234124444CDAB0102034050607177E000000000000"
```

```python
def data_to_chips(hex_str: str) -> List[str]:
    output_array: List[str] = []
    for hex_character in hex_str:
        output_array.append(chip_extended_dict[hex_character])
    return output_array
```

```python
def get_middle(sym1, sym2):
    # i1 = sym1[0]
    i2 = sym2[0]
    q1 = sym1[1]
    # q2 = sym2[1]
    return i2 + q1

```

```python
def combine_chips(chips_list: List[str]) -> List[str]:
    full = " ".join(chips_list)
    symbols = full.split(" ")
    output_array: List[str] = [symbols[0]]
    for idx, symbol in enumerate(symbols[1:]):
        output_array.append(get_middle(symbols[idx], symbol))
        output_array.append(symbol)
    return output_array
```

```python
waveforms_dict = {
    "01": "000000000000111111111111",
    "00": "000000111111111111000000",
    "10": "111111111111000000000000",
    "11": "111111000000000000111111",

};
```

```python
def to_wave_forms(symbols: List[str], repeats: int = 1) -> List[str]:
    output_array: List[str] = []
    for symbol in symbols:
        for i in range(repeats):
            output_array.append(waveforms_dict[symbol])
    return output_array
```

```python
print(to_wave_forms(combine_chips(data_to_chips(data_hex)), 4))
```

```python
def to_lengths(waves: List[str]):
    wave_str = "".join(waves)
    output_array: List[int] = []
    if wave_str[0] == "1":
        output_array.append(1)
    else:
        output_array.append(0)
    count = 1
    curr_char = wave_str[0]
    for char in wave_str[1:]:
        if curr_char != char:
            output_array.append(count)
            curr_char = char
            count = 1
        else:
            count += 1
    return output_array
```

```python
print(to_lengths(to_wave_forms(combine_chips(data_to_chips(data_hex)), 4)))
```

```python
def to_bits(lengths: List[int]):
    is_high = lengths[0] == 1
    values = lengths[1:]
    bits = []
    if is_high:
        bits.append(0)
    for l in lengths:
        if l == 4:
            bits += [0]
        if l == 6:
            bits += [1, 0]
        if l == 8:
            bits += [1, 1, 0]
        if l == 12:
            bits += [1, 1, 1, 1, 0]
        if l == 14:
            bits += [1, 1, 1, 1, 1, 0]
        if l == 16:
            bits += [1, 1, 1, 1, 1, 1, 0]
        if l == 18:
            bits += [1, 1, 1, 1, 1, 1, 1, 0]
        if l == 20:
            bits += [1, 1, 1, 1, 1, 1, 1, 1, 0]
    # print(bits)
    bytes: List[str] = [""]
    idx = 0
    for b in bits:
        if len(bytes[idx]) >= 32:
            idx += 1
            bytes += [""]
        bytes[idx] += str(b)
    while len(bytes[idx]) < 32:
        bytes[idx] += str(0)
    # for s in bytes:
    #     print("0b"+s)
    return bytes
```

```python
bytes = to_bits(to_lengths(to_wave_forms(combine_chips(data_to_chips(data_hex)), 3)))
```
